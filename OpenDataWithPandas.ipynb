```python
df_all = pd.concat([df_a,df_b,df_e,df_f,df_h])
df_all = df_all.fillna(0)
df_all
```

```python
def strToNum(strNum):
    num = 0
    if not strNum:
        return num
    if strNum==0:
        return num
    strNum = strNum.replace('層','')
    switcher = {
        '一':'1',
        '二':'2',
        '三':'3',
        '四':'4',
        '五':'5',
        '六':'6',
        '七':'7',
        '八':'8',
        '九':'9',
        '十':'10'
    }
    numStr ='0' 
    for strs in strNum:
        numStr += switcher.get(strs,'')
    num = int(numStr)
    return num

strFloor = ('')
numFloor = strToNum(strFloor)
print(numFloor)
```

```python
floor = []
for i in df_all['總樓層數']:
    temp = strToNum(i)
    floor.append(temp)
myFloor = pd.Series(floor)
df_all.insert(12,column='總樓層數Int',value=myFloor)
```
